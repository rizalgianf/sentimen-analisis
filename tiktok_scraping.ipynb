```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Path ke chromedriver (sesuaikan dengan lokasi Anda)
driver_path = "/path/to/chromedriver"  # Ganti dengan path Anda

# Setup Selenium WebDriver (misalnya, dengan Chrome)
options = webdriver.ChromeOptions()
options.add_argument("--headless")  # agar browser tidak terbuka
driver = webdriver.Chrome(executable_path=driver_path, options=options)
```

```python
# URL video TikTok yang ingin di-scrape
url = "https://www.tiktok.com/@username/video/video_id"  # Ganti dengan URL video TikTok yang sesuai

# Buka URL TikTok
driver.get(url)

# Tunggu beberapa detik agar halaman sepenuhnya dimuat
time.sleep(5)
```

```python
# Scroll halaman untuk memuat lebih banyak komentar
body = driver.find_element(By.TAG_NAME, "body")
for _ in range(5):  # Meng-scroll 5 kali
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
```

```python
# Ambil konten halaman menggunakan BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Menemukan semua komentar
comments = soup.find_all("div", class_="tiktok-1yvxqri-DivItemContainer")

# Menyimpan komentar dalam sebuah list
comment_list = []
for comment in comments:
    try:
        username = comment.find("a", class_="tiktok-1f1usl3-AAnchor").text
        text = comment.find("div", class_="tiktok-1lxw9sd-Text").text
        comment_list.append({'username': username, 'comment': text})
    except AttributeError:
        pass  # Beberapa elemen mungkin tidak memiliki komentar
```

```python
# Simpan data komentar ke dalam CSV
df_comments = pd.DataFrame(comment_list)
df_comments.to_csv('tiktok_comments.csv', index=False)

# Tampilkan DataFrame untuk memastikan datanya benar
import ace_tools as tools; tools.display_dataframe_to_user(name="TikTok Comments", dataframe=df_comments)
```

```python
# Tutup browser setelah selesai
driver.quit()
```
